```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

<h1>Exercise 1
Dê uma olhada na documentação da função plot, e veja como modificar o tipo de linha, a cor dela, acrescentar um grid e um título ao seu gráfico e aos eixos.

```python
x = np.linspace(1,20,100)
y = x**2
plt.plot(x,y)
```

```python
x = np.linspace(1,20,100)
y = x**2
plt.plot(x,y,'r')
```

```python
x = np.linspace(1,20,100)
y = x**2
plt.plot(x,y,'g',linestyle = '--')
```

```python
x = np.linspace(1,20,100)
y = x**2
plt.plot(x,y,'g',linestyle = '-.')
```

```python
plt.grid()
plt.title("TITULO ERICK")
plt.xlabel('EIXO X ERICK')
plt.ylabel("EIXO Y ERICK")
plt.plot(x,y,'g',linestyle = '--')
```

<h1>Exercise 2
As configurações feitas no exercício anterior podem ser usadas também no método plot da pandas?

Sim, mas as funções de plot do pandas são diferentes.

```python
ts = pd.Series(y)
ts.plot(grid = True,kind='line', title='TITULO ERICK', legend=True, color='green', animated=True)
```

```python
ts = pd.Series(y)
plt.xlabel('EIXO X ERICK')
plt.ylabel('EIXO Y ERICK')
ts.plot(grid = True,kind='line', title='TITULO ERICK', legend=True, color='red', animated=True, linewidth=10)
```
